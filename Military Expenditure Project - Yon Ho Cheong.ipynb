```python
# Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import ExcelWriter
from IPython.display import display
```

```python
# Use 'Military expenditure dataset' as a Data Source and choose the 'current USD' work sheet in Military Expenditure dataset
df = pd.read_excel('Military expenditure dataset.xls', sheetname='Current USD',skiprows=5, skipfooter=8, index_col=0)
df.head(10)
```

```python
# Replace the non-numeric value with zero in a table
df = df.replace('xxx',0)
df = df.replace('. .',0)
df = df.replace('...',0)
df.head(10)
```

```python
# Sort the data on the basis of top ten countries of Military Expenditures
df_7=df.loc[:, [2010,2011,2012,2013,2014,2015,2016]]
df_7.columns=['2010','2011','2012','2013','2014','2015','2016']
df_7=df_7.sort_values(by='2013', ascending=False).head(10)
df_7
```

```python
# Function to plot per capita data
def plot_percapita(df, limit=10):
    df = df.loc[:,'2013']
    df.plot(kind='bar', stacked=True, sort_columns=True)
    plt.show()
```

```python
 plot_percapita(df_7)
```

```python
# Pie chart of Military Expenditure -- Comparing year of 2010 and 2016
plt.figure(figsize=(9,9))
df_7['2010'].plot.pie(subplots=True, legend=True, fontsize=10, startangle=170, autopct='%1.2f%%')
plt.show()

plt.figure(figsize=(9,9))
df_7['2016'].plot.pie(subplots=True, legend=True, fontsize=10, startangle=170, autopct='%1.2f%%')
plt.show()
```

```python
# The fastest growing countries in Military Expenditure: 'China' is the fastest growing country
df_7['2016']-df_7['2010']
```

```python
# Fixed Value in percentages 
fixed_value=(df_7['2016']-df_7['2010'])/df_7['2010']*100
fixed_value
```

```python
# Plot for Fixed Value in Military Expenditure
fixed_value=fixed_value.sort_values(ascending=True)
fixed_value.plot(kind='barh', title='Fixed Value for Military Expenditure')
plt.show()
```

```python
# Get Share of GDP data to find top 10 countries
share_GDP = pd.read_excel('Military expenditure dataset.xls', sheetname='Share of GDP',skiprows=5, skipfooter=8, index_col=0)
share_GDP = share_GDP.replace('xxx',0)
share_GDP = share_GDP.replace('. .',0)
share_GDP = share_GDP.replace('...',0)
share_GDP.head(10)
```

```python
# Indexing the top ten countries in the Share of GDP dataset
Top10_share_GDP= pd.DataFrame(share_GDP, index = ['USA','China, P.R.','Russian Federation','Saudi Arabia','France','UK','Japan','India','Germany','Korea, South'])
Top10_share_GDP
```

```python
# Find the top ten countries during 7 years
Top10_share_GDP = Top10_share_GDP.loc[:, [2010,2011,2012,2013,2014,2015,2016]]
Top10_share_GDP.columns=['2010','2011','2012','2013','2014','2015','2016']
Top10_share_GDP = Top10_share_GDP.sort_values(by='2013',ascending=False).head(10)
Top10_share_GDP
```

```python
# Military expenditure by country as percentage of gross domestic product(Graph)
plt.figure(figsize=(16,10))
for i in Top10_share_GDP.columns:
    plt.plot(range(len(Top10_share_GDP[i].index)),Top10_share_GDP[i].values)
plt.xticks(range(10),Top10_share_GDP.index, rotation=30)
plt.ylabel('Share of GDP')
plt.title('Top 10 Countries Shares of GDP in Military Expenditure ')
plt.legend(Top10_share_GDP.columns)
plt.grid(True)
plt.show()
```

```python
# line chart for top 10 countries
Top10_share_GDPTransposed = Top10_share_GDP.transpose()
Top10_share_GDPTransposedPlot = Top10_share_GDPTransposed.plot.line(legend=False,figsize=(12,8),fontsize=10)
Top10_share_GDPTransposedPlot.set_ylabel('Share of GDP',fontsize=10)
Top10_share_GDPTransposedPlot.set_xlabel('Year',fontsize=10)
Top10_share_GDPTransposedPlot.set_title('Top 10 Countries in Military Expenditure',fontsize=15)
Top10_share_GDPTransposedPlot.legend()
```

```python
# Use GDP Per Capita as a Data Source and replace the non-numeric value with zero in a table
GDP_per_capita = pd.read_excel('GDP per capita.xls', sheetname='Data',skiprows=3, index_col=0)
GDP_per_capita = GDP_per_capita.replace('xxx',0)
GDP_per_capita = GDP_per_capita.replace('. .',0)
GDP_per_capita = GDP_per_capita.replace('...',0)
GDP_per_capita.head()
```

```python
# Find the year of 2010 to 2016
GDP_per_capita = GDP_per_capita.loc[:, ['2010','2011','2012','2013','2014','2015','2016']]
GDP_per_capita.head(10)
```

```python
# Change the column's name
GDP_per_capita.columns=['2010_GDP_per_capita','2011_GDP_per_capita','2012_GDP_per_capita','2013_GDP_per_capita','2014_GDP_per_capita','2015_GDP_per_capita','2016_GDP_per_capita']
GDP_per_capita.head()
```

```python
# Rename the countries which are not matched
GDP_per_capita=GDP_per_capita.rename(index={'United States': 'USA'})
GDP_per_capita=GDP_per_capita.rename(index={'China': 'China, P.R.'})
GDP_per_capita=GDP_per_capita.rename(index={'United Kingdom': 'UK'})
GDP_per_capita=GDP_per_capita.rename(index={'Korea, Rep.': 'Korea, South'})

Top10_GDP_per_capita = pd.DataFrame(GDP_per_capita, index = ['USA','China, P.R.','Russian Federation','Saudi Arabia','India','France','UK','Japan','Germany','Korea, South'])
Top10_GDP_per_capita
```

```python
# Add 'Average' column
Top10_GDP_per_capita['Average'] = Top10_GDP_per_capita.mean(axis=1)
Top10_GDP_per_capita
```

```python
# Pie chart for Average of GDP per capita
plt.figure(figsize=(9,9))
Top10_GDP_per_capita['Average'].plot.pie(subplots=True, legend=False, fontsize=10, startangle=170, autopct='%1.2f%%')
plt.show()
```

```python
# Use Military Expenditure per capita as a Data Source and replace the non-numeric value with zero in a table
ME_per_capita = pd.read_excel('Military expenditure dataset.xls', sheetname='Per capita',skiprows=6, skipfooter=8, index_col=0)
ME_per_capita = ME_per_capita.replace('xxx',0)
ME_per_capita = ME_per_capita.replace('. .',0)
ME_per_capita = ME_per_capita.replace('...',0)
ME_per_capita.head()
```

```python
# Find the 7 years data for Military Expenditure Per capita dataset
ME_per_capita = ME_per_capita.loc[:, [2010,2011,2012,2013,2014,2015,2016]]
ME_per_capita.head()
```

```python
# Rename the columns
ME_per_capita.columns=['2010_ME_per_capita','2011_ME_per_capita','2012_ME_per_capita','2013_ME_per_capita','2014_ME_per_capita','2015_ME_per_capita','2016_ME_per_capita']
ME_per_capita.head()
```

```python
# Rename the countries and find top 10 countries' data
ME_per_capita=ME_per_capita.rename(index={'USSR/Russia': 'Russian Federation'})
Top10ME_per_capita = pd.DataFrame(ME_per_capita, index = ['USA','China, P.R.','Russian Federation','Saudi Arabia','India','France','UK','Japan','Germany','Korea, South'])
Top10ME_per_capita
```

```python
# Merge two tables together
ME_GDP=Top10ME_per_capita.join(Top10_GDP_per_capita)
ME_GDP
```

```python
# Compare the per person military spending to the per person GDP -
ME_GDP['2016_ME_per_capita']/ME_GDP['2016_GDP_per_capita']
```

```python
# per person military spending to the per person GDP

Growing = ME_GDP['2016_ME_per_capita']/ME_GDP['2016_GDP_per_capita']

Growing.plot.bar(fontsize=10, legend=False, figsize=(12,8))

plt.title('Per Person Military Spending to the Per Person GDP')

plt.show()
```
